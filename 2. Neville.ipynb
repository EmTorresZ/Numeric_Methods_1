```python
import numpy as np
from numpy.polynomial.polynomial import Polynomial
import matplotlib.pyplot as plt
```

```python
# Obtain how many data we have, and make the subindex list
print("Please enter the number of data")
n = int(input())
Subindex = list(range(n))
```

```python
# Select Data and display it in coordenates
x_ = []
y_ = []
for i in Subindex :
  print("Please enter the value for x_", i)
  x_.append(float(input()))
  print("Please enter the value for y_", i)
  y_.append(float(input()))
  
plt.scatter(x_,y_)
plt.show()
```

```python
# Set the P polynomals
P = [[0.0] * (r+1) for r in Subindex]
row = [0.0] * n # row of the iteration table
for i in Subindex :
   row[i] = str(i) + "\t" + str(x_[i])
   for j in Subindex :
      if j <= i :
         if j == 0 :
           P[i][0] = Polynomial(coef=[x_[i]])
         else :
            P[i][j] = (Polynomial(coef=[-x_[i-j], 1])*P[i][j-1] - Polynomial(coef=[-x_[i], 1])*P[i-1][j-1]) / (x_[i] - x_[i-j])
         row[i] += "\t" + str(P[i][j])
```

```python
# Iteration Table
table_tittle = "\ni \t x_i \t y_i"    
for j in Subindex :
   if j > 0 :
      table_tittle += "\t" + str(j) # set the table tittle
print(table_tittle, "\n")
for i in Subindex :
   print(row[i])
```

```python
# We show the graphic using z instead of x
y = P[n-1][n-1]
z = np.linspace(start=x_[0], stop=x_[n-1], num=(n+1))
plt.scatter(x_,y_)
plt.plot(z, y(z))
plt.show()

print("y =", y)
```

```python
# Value for the Interpolation
print("Please enter the number you want to make the interpolation for")
x = float(input())
```

```python
# We show the graphic using z instead of x
z = np.linspace(start=min(x_[0], x), stop=max(x_[n-1], x), num=(n+2))
plt.scatter(x_,y_)
plt.plot(z, y(z))
plt.scatter(x,y(x))
plt.show()

print("y =", y)
print("y(", x, ") =", y(x))
```
